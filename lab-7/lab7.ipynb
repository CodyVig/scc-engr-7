Now, write code to find the <b>most</b> frequently occuring number in this updated version `z` (`z` should contain no negative numbers).

This can be done either with loop, or using a NumPy functions. Either way is fine.

<b>np.bincount</b> - Count number of occurrences of each value in array of non-negative ints.


````python
# Using NumPy
bins = np.bincount(z)
most_frequent = np.where(bins == max(bins))[0][0]

print(f"[Using bincount()] The most frequently occuring number is {most_frequent}.")

# Using dictionaries, for fun
occurences = {}
for value in z:
    if value not in occurences:
        occurences[value] = 1
    else:
        occurences[value] += 1

most_frequent = max(occurences, key=occurences.get)
print(f"[Using dictionary] The most frequently occuring number is {most_frequent}.")
````

<b> (20 points)</b> Choose  one of the physiological variable from the above (pulse, weight, or waist). Make a plot with 3 subplots of the physiological variable (y-axis) vs each of the the exercise variable (x-axis). Add an appropriate plot title, xlabel, and ylabel. Add a legend and gridlines. 

Here is an example for the "pulse" physiological variable



matplotlib functions to use
* plt.subplot
* plt.figure
* plt.title
* plt.xlabel
* plt.ylabel
* plt.grid
* plt.legend

Use plt.plot() inputs like the marker style, color, linewidth, and marker size.

````python
def plot_physiological_variable(
        variable: np.array, 
        units: str,
        include_jumps: bool = True, 
        include_chinups: bool = True, 
        include_situps: bool = True
    ):
    """
    Plots `variable` against `jumps`, `chinups`, `situps` data. 

    :param variable: `np.array` representing either weight, pulse, or waist.
    :param units: The y-axis label for the plots.
    """

    number_of_figs = include_jumps + include_chinups + include_situps
    fig_number = 0

    if number_of_figs == 0:
        print("Nothing to plot.")
        return

    plt.rcParams["figure.figsize"] = [4 * number_of_figs, 4]
    plt.rcParams["figure.autolayout"] = True

    if include_jumps:
        fig_number += 1
        plt.subplot(1, number_of_figs, fig_number)
        plt.scatter(
            jumps, 
            variable, 
            marker="o", 
            alpha = 0.5, 
            color = 'c', 
            label=units.split()[0]
        )
        plt.xlabel("Jumps")
        plt.ylabel(units)
        plt.title(f"{units.split()[0]} vs Jumps")
        plt.grid()
        plt.legend()
    
    if include_chinups:
        fig_number += 1
        plt.subplot(1, number_of_figs, fig_number)
        plt.scatter(
            chinups, 
            variable, 
            marker=">", 
            alpha = 1, 
            color = 'r', 
            label=units.split()[0]
        )
        plt.xlabel("Chinups")
        plt.ylabel(units)
        plt.title(f"{units.split()[0]} vs Chinups")
        plt.grid()
        plt.legend()

    if include_situps:
        fig_number += 1
        plt.subplot(1, number_of_figs, fig_number)
        plt.scatter(
            situps, 
            variable, 
            marker="*", 
            alpha = 1, 
            color = 'y', 
            label=units.split()[0]
        )
        plt.xlabel("Situps")
        plt.ylabel(units)
        plt.title(f"{units.split()[0]} vs Situps")
        plt.grid()
        plt.legend()

    plt.show()
````

````python
plot_physiological_variable(variable=weight, units="Weight (lbs)")
````

````python
plot_physiological_variable(variable=pulse, units="Pulse (bpm)")
````

````python
plot_physiological_variable(variable=waist, units="Waist (in)")
````

````python
def bisection(f, a, b, epsilon):
    """Find a root of a function f beetween a and b within 
    epsilon accuracy using the bisection method.
    
    Inputs
        f - A mathematical function
        a - a number with  f(a) < 0
        b - a number with f(b) > 0
        epsilon - A tolerance
    """

    # Check if there is a root between a and b -- f(a) < 0 and f(b) > 0
    if not (f(a) <= 0 and f(b) >= 0): 
        print("There is no root between", a , "and", b, ".")
        return
    
    m = (a + b) / 2
    
    if abs(f(m)) <= epsilon:
        return m
    
    elif f(m) < 0:
        return bisection(f, m, b, epsilon)
    
    else:
        return bisection(f, a, m, epsilon)
````

````python
"""
For fun, I wrote a version of `bisection` that works on decreasing functions.
"""

from typing import Callable


def new_bisection(
    f: Callable[[float], float], 
    a: float, 
    b: float, 
    epsilon: float = 10 ** (-6),
):
    """
    Runs the bisection algorithm until a root of `f` in [`a`, `b`] is found 
    within tolerance `epsilon`.
    """

    m = a
    while abs(f(m)) > epsilon:
        m = do_bisect(f, a, b)
        fm = f(m)

        if f(a) * fm < 0:
            b = m
        elif fm * f(b) < 0:
            a = m
        else:
            return m
    
    return m


def do_bisect(f: Callable[[float], float], a: float, b: float) -> float:
    """
    Helper function to run one iteration of bisection.
    """

    if f(a) * f(b) > 0:
        raise ValueError(f"No roots in [{a}, {b}].")
    if f(a) == 0:
        return a
    if f(b) == 0:
        return b
    return (a + b) / 2
````
